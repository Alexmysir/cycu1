```python
num = int(input("請輸入乘法表的大小："))  # 取得使用者輸入的數字，作為乘法表的大小
for i in range(1, num + 1):  # 外層迴圈，i 從 1 到 num，代表乘法表的列
    for j in range(1, num + 1):  # 內層迴圈，j 從 1 到 num，代表乘法表的欄
        mul = i * j  # 計算 i 和 j 的乘積
        print(f"{i} x {j} = {mul}", end=" ")  # 輸出乘積結果，使用 end=" " 讓輸出保持在同一行
    print()  # 換行，開始下一列的輸出
```

```python
from google.colab import files
import pandas as pd
import io

uploaded = files.upload()

# 正確從 dict 取出第一個上傳檔案
filename, filecontent = next(iter(uploaded.items()))

# 根據副檔名選擇合適讀法
if filename.lower().endswith(('.xls', '.xlsx')):
    df = pd.read_excel(io.BytesIO(filecontent))
elif filename.lower().endswith('.csv'):
    df = pd.read_csv(io.StringIO(filecontent.decode('UTF-8')))
else:
    raise ValueError(f"不支援的檔案格式：{filename}")


df['合計'] = df['北區'] + df['中區'] + df['南區']
grouped = df.groupby('業務')['合計'].sum().reset_index()
# 依照合計欄排序取前 3 名
top3 = grouped.sort_values(by='合計', ascending=False).head(3)
print("合計前三名業務如下：")
print(top3)

```

```python
from google.colab import files
import pandas as pd
import io
from pathlib import Path

# ─── 步驟一：上傳並讀取檔案 ───
uploaded = files.upload()
orig_filename, content = next(iter(uploaded.items()))

# 判斷副檔名
suffix = orig_filename.lower().rsplit(".", 1)
if len(suffix) != 2:
    raise ValueError("檔名沒有副檔名，無法自動辨認格式：%s" % orig_filename)
base, ext = suffix
ext = ext.lower()

# 建議 Excel 使用 BytesIO，CSV 使用 StringIO
if ext in ("xls", "xlsx"):
    df = pd.read_excel(io.BytesIO(content))
elif ext == "csv":
    df = pd.read_csv(io.StringIO(content.decode("utf-8")))
else:
    raise ValueError(f"不支援檔案格式：{ext}")

# ─── 步驟二：執行資料處理（範例中移除「北區」欄） ───
mydata = df.drop(columns=["北區"], errors="ignore")

# ─── 步驟三：指定新的檔名，加上 _1 後綴 ───
new_suffix = base + "_1." + ext
#⁣ 例如原始 '002.xlsx' → '002_1.xlsx'

# ─── 步驟四：save 到 Colab VM，再使用瀏覽器下載到本機 ───
if ext in ("xls", "xlsx"):
    # Excel 格式：需安裝 openpyxl 或 xlsxwriter 等
    mydata.to_excel(new_suffix, index=False)
else:
    # CSV 格式
    mydata.to_csv(new_suffix, index=False, encoding="utf‑8-sig")

# ─── 最後一步：觸發瀏覽器下載流程 ───
files.download(new_suffix)
print(f"已將處理後的檔案包成：{new_suffix}，下載對話框已彈出。")
```
